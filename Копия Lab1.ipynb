```python
def to_date(x):
  splited = x.split('-')
  new_date = splited[-1].zfill(2) + '-' + splited[-2].zfill(2)+ '-' + splited[-3]
  return new_date

df['CET'] = df['CET'].apply(to_date)
df['CET'].head()
```

```python
day_df = df.copy()
day_df['Date'] = day_df['CET'].apply(lambda x: '-'.join(x.split('-')[0:2]))
day_df.groupby('Date')['Max TemperatureC'].mean()
```

```python
day_df['Year'] = pd.DatetimeIndex(day_df['CET']).year
foggy_days_per_year = day_df[day_df[' Events'].str.contains('Fog', na=False)].groupby('Year').size()
print("Number of foggy days each year:")
print(foggy_days_per_year)
```

```python
event_counts = df[' Events'].value_counts()
plt.figure(figsize=(10, 5))
event_counts.plot(kind='bar')
plt.title('Number of Events')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

9

```python
wind_directions = df['WindDirDegrees']

def wind_direction(degree):
    if degree >= 337.5 or degree < 22.5:
        return 'North'
    elif 22.5 <= degree < 67.5:
        return 'North-East'
    elif 67.5 <= degree < 112.5:
        return 'East'
    elif 112.5 <= degree < 157.5:
        return 'South-East'
    elif 157.5 <= degree < 202.5:
        return 'South'
    elif 202.5 <= degree < 247.5:
        return 'South-West'
    elif 247.5 <= degree < 292.5:
        return 'West'
    elif 292.5 <= degree < 337.5:
        return 'North-West'

df['Wind Direction'] = df['WindDirDegrees'].apply(wind_direction)
wind_direction_counts = df['Wind Direction'].value_counts()
plt.figure(figsize=(8, 8))
wind_direction_counts.plot(kind='pie')
plt.title('Wind Directions')
plt.ylabel('')
plt.show()
```

10

```python
day_df['Month'] = pd.DatetimeIndex(day_df['CET']).month
monthly_avg_max_temp = day_df.groupby(['Year', 'Month'])['Max TemperatureC'].mean()
monthly_avg_min_dew_point = day_df.groupby(['Year', 'Month'])['Min DewpointC'].mean()

plt.figure(figsize=(14, 7))
monthly_avg_max_temp.plot(label='Average Max Temperature')
monthly_avg_min_dew_point.plot(label='Average Min Dew Point')
plt.title('Monthly Average Max Temperature and Min Dew Point')
plt.xlabel('Year-Month')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
